```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
x_square = x**2
```

```python
x
```

```python
x_square
```

```python
plt.plot(x, x, marker ="*", color="red", label ="y=x")
plt.plot(x, x_square, marker = "o" , label = "y = x**2")
plt.xlabel("x axis")
plt.ylabel("Graphs")
plt.legend()
plt.title("Comparison Graph")
plt.show()
```
