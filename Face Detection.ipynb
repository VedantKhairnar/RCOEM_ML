```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread("images.jpeg")
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
face_detector = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = face_detector.detectMultiScale(img , 1.05, 1)
```

```python
faces
```

```python
for face in faces:
    x,y,w,h = face
    rect_img = cv2.rectangle(img, (x, y), (x+w, y+h) , (255,255,255), 2)

plt.imshow(rect_img)    
```

```python
%matplotlib
```

```python
cam = cv2.VideoCapture(0)
face_detector = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
while True:
    ret, frame = cam.read()
    if ret == False:
        continue
    
    faces = face_detector.detectMultiScale(frame , 1.05, 10)
    for face in faces:
        x,y,w,h = face
        rect_img = cv2.rectangle(frame, (x, y), (x+w, y+h) , (255,255,255), 2)
    
    cv2.imshow("Face Frame", frame)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break
        
cam.release()
cv2.destroyAllWindows()
```
